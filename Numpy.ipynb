# Numpy Arrays

NumPy arrays are the main way we will use Numpy throughout the course. Numpy arrays essentially come in two flavors: vectors and matrices. Vectors are strictly 1-d arrays and matrices are 2-d (but you should note a matrix can still have only one row or one column).

Let's begin our introduction by exploring how to create NumPy arrays.

## Creating NumPy Arrays

### From a Python List

We can create an array by directly converting a list or list of lists:

```python
a = [1,2,3,4,5] # Its is just a normal list
b = ["1",2,3,4,5]  # saare element ka datatype change kar deta hai
arr = np.array(a)   # converting the list to a numpy array using -> np.array()
arr1 = np.array(b)
print(arr)
print(arr1)
```

Lets create a matrix using list

```python
a = [[1,2,3],[4,5,6],[7,8,9]]
a
```

Now lets see how matlix looks like using numpy arrays

```python
arr = np.array(a)
arr
```

## Built-in Methods

there are lots of build in methods to create an array lets create some and see

np.zeros(), np.ones(), np.full()

np.arange(), np.linspace()

np.eye()

```python
np.zeros(4)
```

```python
np.zeros([3,4]) #row and columns
```

```python
np.ones(3)
```

```python
np.ones([2,2])
```

```python
np.full(5,7)
```

```python
np.full([5,5],7)
```

```python
# arange use for arranging an array
arr = np.arange(0, 10, 2)   # start=0, stop=10, step=2
print(arr)
```

```python
# linspace -> create an array between value 0 to 1 and usme 5 values honi chaiye orr 5 values ke bech mai same difference (linespace) ho.
arr = np.linspace(0, 1, 5)   # start=0, stop=1, 5 values
print(arr)
```

```python
# eye -> 4x4 identity matrix
arr = np.eye(4)
print(arr)
```

```python
# random.rand -> give a random number between 0 to 1
arr = np.random.rand(2, 3)   # 2x3 array
print(arr)
```

```python
# random.randint -> give a random number between your provided range (1 to 10 , size(3, 3))
arr = np.random.randint(1, 10, [3, 3])   # 3x3 integers from 1–9
print(arr)
```

```python
# random.randn(size) -> give you a random number between -3 to +3
arr = np.random.randn(2, 4)   # 2x4 array
print(arr)
```

arange → step-based values

linspace → fixed number of values

eye → identity matrix

rand → random floats (0–1)

randint → random integers in a range

randn → random values from bell curve distribution
