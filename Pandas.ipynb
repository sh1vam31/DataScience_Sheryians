# Class 2: DataFrame Basics
Welcome to **Class 2**. Today we’ll learn how to **inspect**, **select**, and **organize** data inside a DataFrame.

## 1 Inspecting Data
Quick ways to understand your dataset:
- `head(n)` → first *n* rows (default 5)
- `tail(n)` → last *n* rows
- `info()` → columns, non-null counts, and dtypes
- `shape` → (rows, columns)
- `dtypes` → data type of each column

````python
import pandas as pd

# Create a small DataFrame
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)

# 1. See the first rows
print("First 3 rows:")
print(b.head(3))

# 2. See the last rows
print("\nLast 2 rows:")
print(b.tail(2))

# 3. Get a concise summary
print("\nInfo about DataFrame:")
print(b.info())

# 4. Shape of the DataFrame (rows, columns)
print("\nShape:")
print(b.shape)

# 5. Data types of each column
print("\nData types:")
print(b.dtypes)

````

## 2 Selecting Columns & Rows
Ways to access parts of a DataFrame:
- Single column → `df['col']`
- Multiple columns → `df[['col1','col2']]`
- Label-based selection → `df.loc[row_label, col_label]`
- Integer-index selection → `df.iloc[row_idx, col_idx]`
Tip: `:` means "all rows/columns" in that dimension.

````python

# 1. Single column
print("Single column (Name):")
print(b["Name"])

# 2. Multiple columns
print("\nMultiple columns (Name & Score):")
print(b[["Name", "Score"]])

# 3. Label-based selection with .loc ->
print("\nLabel-based selection (row index 2, column 'Score'):")
print(b.loc[2, "Score"])

print("\nRows 1 to 3 with columns Name & Age:")
print(b.loc[1:3, ["Name", "Age"]])

# 4. Integer-index selection with .iloc
print("\nRow at index 3 (all columns):")
print(b.iloc[3])

print("\nRows 0 to 2, columns 0 to 1:")
print(b.iloc[0:3, 0:2])
````

## 3 Index & Columns Overview
- `df.index` → gives row name or number
- `df.columns` → gives column names
- `df.values` → underlying NumPy array (read-only; prefer using DataFrame ops)

````python

# Row labels (index)
print("Index:")
print(b.index)

# Column names
print("\nColumns:")
print(b.columns)

# Underlying NumPy array
print("\nValues:")
print(b.values)

# Optional: shape of the underlying array
print("\nShape of values:")
print(b.values.shape)
````

## 4 Renaming Columns & Setting an Index
- Rename columns → `df.rename(columns={'old':'new'}, inplace=False)` -> inplace = True (chnages made in original data)
- Set a column as index → `df.set_index('column_name', inplace=False)` -> inplace = False (chnages made only in the copy of the data)
- Reset index → `df.reset_index(inplace=False)`
 If you do inplace = True by mistace and want the original data back so use this


````python

# 1. Rename columns
c = b.rename(columns={"Score": "Marks"})
print("After renaming 'Score' to 'Marks':")
print(c)

# 2. Set a column as index
d = c.set_index("Name")
print("\nAfter setting 'Name' as index:")
print(d)

# 3. Reset the index back to default
e = d.reset_index()
print("\nAfter resetting the index:")
print(e)
````

## 5 Basic Attributes Recap
- `df.columns` → names of all columns
- `df.index` → row labels
- `df.values` → NumPy array (data only)
These are handy for quick checks and loops (though vectorized ops are preferred).

# Class 3: Data Cleaning Essentials
We’ll learn how to clean and prepare data for analysis.

## Handling Missing Values
Real-world datasets often have missing data. Pandas offers:
- `isna()` / `isnull()` → detect missing values (returns True/False)
- `dropna()` → remove missing values (rows/columns)
- `fillna()` → replace missing values with a given value or method (e.g., forward fill)

Always inspect how much data is missing before deciding to drop or fill.

````python
import pandas as pd
import numpy as np

# Create a sample DataFrame with some missing values
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, None, 35, 40, None],
    "City": ["Delhi", "Mumbai", None, "Bangalore", "Delhi"],
    "Score": [85, 91, np.nan, 88, 95]
}

b = pd.DataFrame(a)
print("Original DataFrame:")
print(b)

# 1. Detect missing values
print("\nCheck where data is missing:")
print(b.isna())

print("\nCount of missing values in each column:")
print(b.isna().sum())

# 2. Drop rows with any missing value
c = b.dropna() # it will drop full row jisme akkk bhi missing(Nan) value hogi
print("\nAfter dropping rows with missing values:")
print(c)

# 3. Fill missing values
# Fill numeric columns with mean, text columns with a placeholder
d = b.copy()
d["Age"] = d["Age"].fillna(d["Age"].mean())
d["City"] = d["City"].fillna(d["City"].mode()[0]) # accessing the first element of the mode Series
d["Score"] = d["Score"].fillna(d['Score'].mean())
print("\nAfter filling missing values:")
print(d)

print(b.isna().sum())   # show old data becouse we did not do inplac = true
````

## Changing Data Types (`astype`)
- Convert a column to a different type using `astype()`.
- Examples: converting strings to numbers, integers to floats, or columns to category.

Be careful: make sure the data is compatible with the new type!

````python
import pandas as pd

# Same dataset
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
print("Original DataFrame:")
print(b)

# 1. Convert "Age" column from int to float
b["Score"] = b["Score"].astype(float)
print("\nScore column converted to float:")
print(b.dtypes) # dtype -> use for checking the datatype of the column

# 2. Convert "Name" column to category type (memory-efficient)
b["Name"] = b["Name"].astype("category")
print("\nName column converted to category:")
print(b.dtypes)

# 3. Convert "Score" column to string
b["Score"] = b["Score"].astype(str)
print("\nScore column converted to string:")
print(b.dtypes)
````

## Replacing Values (`replace`)
- Use `replace()` to substitute specific values or patterns.
- Useful for cleaning labels, fixing typos, or mapping categories.

````python

# 1. Replace a single value in a column
c = b.copy()
c["Name"] = c["Name"].replace("Bob", "Robert")
print("\nAfter replacing 'Bob' with 'Robert':")
print(c)

# 2. Replace multiple values at once
d = b.copy()
d["Score"] = d["Score"].astype(float)
d["Score"] = d["Score"].replace({88.0: 90.0, 95.0: 99.0})
print("\nAfter replacing 88 -> 90 and 95 -> 99 in Score:")
print(d)

# 3. Replace using a pattern in strings (e.g., shorten names)
e = b.copy()
e["Name"] = e["Name"].replace({"Charlie": "Ch"}, regex=False)
print("\nAfter shortening 'Charlie' to 'Ch':")
print(e)
````

## Dropping Rows & Columns (`drop`)
- `drop(labels, axis=0)` → drop rows by index labels.
- `drop(labels, axis=1)` → drop columns by name.
- Use `inplace=True` if you want to modify the DataFrame directly.

````python

# 1. Drop a single row by index label
c = b.drop(2)   # drops row with index 2 (Charlie)
print("\nAfter dropping row with index 2:")
print(c)

# 2. Drop multiple rows
d = b.drop([0, 4])   # drops Alice and Emma
print("\nAfter dropping rows 0 and 4:")
print(d)

# 3. Drop a single column
e = b.drop("Score", axis=1) # axis =1 matlab vo column mai Score ko dhoonde ga orr drop karega
print("\nAfter dropping the 'Score' column:")
print(e)

# 4. Drop multiple columns
f = b.drop(["Age", "Score"], axis=1)
print("\nAfter dropping 'Age' and 'Score' columns:")
print(f)

# 5. Using inplace=True to modify the DataFrame directly
g = b.copy()
g.drop("Score", axis=1, inplace=True)
print("\nAfter dropping 'Score' with inplace=True:")
print(g)
````

## Detecting & Removing Duplicates
- `duplicated()` → returns a Boolean Series marking duplicate rows.
- `drop_duplicates()` → remove duplicate rows.
- You can specify `subset` (columns) and `keep` ('first', 'last', or False).

````python
import pandas as pd

# Create a custom dataset with some duplicate rows
data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "Bob", "David"],
    "Age": [25, 30, 35, 25, 30, 40],
    "Score": [88, 92, 79, 88, 92, 85]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# 1. Detect duplicates
print("\nBoolean Series showing duplicates:")
print(df.duplicated()) # .duplicated()-> gives you the rows that have the duplicated values

# 2. Count total duplicates
print("\nTotal number of duplicate rows:")
print(df.duplicated().sum()) # .duplicated().sum() ->  gives you the sum of  rows that have the duplicated values

# 3. Remove duplicates (keep first occurrence)
df_first = df.drop_duplicates()
print("\nAfter dropping duplicates (keep='first'):")
print(df_first)

# 4. Remove duplicates (keep last occurrence)
df_last = df.drop_duplicates(keep="last") # drop the duplicated row but keep the last duplicated row
print("\nAfter dropping duplicates (keep='last'):")
print(df_last)

# 5. Remove all duplicates (keep=False)
df_none = df.drop_duplicates(keep=False) # drop the duplicated row orr kuch bhi keep mat karo
print("\nAfter removing all duplicates (keep=False):")
print(df_none)

# 6. Remove duplicates based on a subset of columns (e.g., Name only)
df_subset = df.drop_duplicates(subset=["Name"]) # Only consider the 'Name' column when deciding if a row is a duplicate
print("\nAfter dropping duplicates based on 'Name' only:")
print(df_subset)
````

# Class 4: Data Selection & Filtering
Welcome to **Class 4** of our Pandas series. Today we’ll explore techniques to select and filter data in a DataFrame.

## 1. Boolean Indexing & Conditional Selection
- Use comparison operators to build conditions (e.g., `df['col'] > 50`).
- Pass the condition inside `df[...]` to get matching rows.
- Combine multiple conditions with `&` (AND), `|` (OR), and `~` (NOT).
- Parentheses are required when combining conditions.

````python
import pandas as pd
import numpy as np
# Same dataset
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)
print("Original DataFrame:")
print(b)

# 1. Single condition: rows where Age > 30
print("\nRows where Age > 30:")
print(b[b["Age"] > 30])

# 2. Single condition: rows where Score < 90
print("\nRows where Score < 90:")
print(b[b["Score"] < 90])

# 3. Multiple conditions (AND): Age > 25 AND Score > 85
print("\nRows where Age > 25 AND Score > 85:")
print(b[(b["Age"] > 25) & (b["Score"] > 85)])

# 4. Multiple conditions (OR): Age < 25 OR Score > 90
print("\nRows where Age < 25 OR Score > 90:")
print(b[(b["Age"] < 25) | (b["Score"] > 90)])

# 5. Using NOT (~): rows where Score is NOT greater than 85
print("\nRows where Score is NOT greater than 85:")
print(b[~(b["Score"] > 85)])
````

## 2. Filtering Rows with Multiple Conditions
- Combine two or more conditions to refine selection.
- Remember to enclose each condition in parentheses.

````python
# 1. Age > 25 AND Score > 85
print("\nRows where Age > 25 AND Score > 85:")
print(b[(b["Age"] > 25) & (b["Score"] > 85)])

# 2. Age < 35 OR Score >= 90
print("\nRows where Age < 35 OR Score >= 90:")
print(b[(b["Age"] < 35) | (b["Score"] >= 90)])

# 3. Age between 25 and 35 (inclusive)
print("\nRows where Age is between 25 and 35 (inclusive):")
print(b[(b["Age"] >= 25) & (b["Age"] <= 35)])
````

## 3. Using `query()` for Cleaner Filtering
- `query()` allows filtering rows with a string expression.
- It's concise and readable for complex conditions.
- Variables can be passed using the `@` symbol.

````python
# 1. Simple query: Age > 25
print("\nRows where Age > 25:")
print(b.query("Age > 25"))  # condition lagane ka naya treeka hai bas

# 2. Multiple conditions: Age < 35 and Score >= 85
print("\nRows where Age < 35 and Score >= 85:")
print(b.query("Age < 35 and Score >= 85"))

# 3. Using a Python variable inside query (with @)
threshold = 90
print(f"\nRows where Score > {threshold}:")
print(b.query("Score > @threshold"))
````

## 4. String Operations on Columns
- Use the `.str` accessor for vectorized string methods.
- Common methods: `contains`, `replace`, `startswith`, `endswith`, `lower`, `upper`.
- Useful for cleaning or filtering text data.

````python
print(b)

# 1. Convert names to lowercase
print("\nNames in lowercase:")
print(b["Name"].str.lower())

# 2. Convert names to uppercase
print("\nNames in uppercase:")
print(b["Name"].str.upper())

# 3. Check if names contain the letter 'a' (case-insensitive)
print("\nRows where Name contains 'a':")
print(b[b["Name"].str.contains("a", case=False)])

# 4. Replace a substring (change 'Charlie' to 'Charles')
c = b.copy()
c["Name"] = c["Name"].str.replace("Charlie", "Charles", regex=False)
print("\nAfter replacing 'Charlie' with 'Charles':")
print(c)

# 5. Filter rows where Name starts with 'D'
print("\nRows where Name starts with 'D':")
print(b[b["Name"].str.startswith("D")])

# 6. Filter rows where Name ends with 'e'
print("\nRows where Name ends with 'e':")
print(b[b["Name"].str.endswith("e")])
````

## Mini Practice
1. Build a DataFrame with columns such as Name, Age, City, Score.
2. Use boolean indexing to select rows where Age > 25.
3. Filter rows where City is either 'Delhi' or 'Mumbai'.
4. Use query() to find rows with Score >= 85 and Age < 35.
5. Clean a text column using str.replace or str.lower.

````python
import pandas as pd

# 1. Build a DataFrame with columns: Name, Age, City, Score
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma", "Frank"],
    "Age": [24, 32, 28, 40, 22, 30],
    "City": ["Delhi", "Mumbai", "Pune", "Chennai", "Delhi", "Mumbai"],
    "Score": [90, 82, 88, 75, 95, 84]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# 2. Use boolean indexing to select rows where Age > 25
age_filter = df[df["Age"] > 25]
print("\nRows where Age > 25:")
print(age_filter)

# 3. Filter rows where City is either 'Delhi' or 'Mumbai'
city_filter = df[df["City"].isin(["Delhi", "Mumbai"])]
print("\nRows where City is Delhi or Mumbai:")
print(city_filter)

# 4. Use query() to find rows with Score >= 85 and Age < 35
query_filter = df.query("Score >= 85 and Age < 35")
print("\nRows with Score >= 85 and Age < 35:")
print(query_filter)

# 5. Clean a text column using str.replace or str.lower
# Example: convert City to lowercase and replace 'delhi' with 'New Delhi'
cleaned = df.copy()
cleaned["City"] = cleaned["City"].str.lower()           # to lowercase
cleaned["City"] = cleaned["City"].str.replace("delhi", "New Delhi", regex=False)
print("\nAfter cleaning City column:")
print(cleaned)
````
